```python
import pandas as pd
from pathlib import Path

# Chemin vers le fichier Parquet
file_path = Path("data/gold/Device_20250416.parquet")

# Lire le fichier Parquet
df = pd.read_parquet(file_path)

# Filtrer les lignes avec un ResellerId non vide ou non nul
filtered_df = df[df['ResellerId'].notna() & (df['ResellerId'] != '')]

# Afficher un aperçu des données filtrées
print(filtered_df.head())

# Afficher le nombre de lignes avec un ResellerId
print(f"\nNombre de lignes avec ResellerId : {len(filtered_df)}")
```
